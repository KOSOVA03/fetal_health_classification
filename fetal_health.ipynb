```python
#prolongued_decelerations
import matplotlib.pyplot as plt
import seaborn as sns
prolongued_decelerations = df.groupby("prolongued_decelerations")["fetal_health"].sum().sort_values(ascending = False)
prolongued_decelerations_df = pd.DataFrame(list(prolongued_decelerations.items()), columns=['prolongued_decelerations', 'fetal_health'])
plt.figure(figsize=(5,5))
plt.title('uzun sÃ¼reli yavaÅŸlamalar ile fetal saÄŸlÄ±ÄŸÄ± oranÄ±')
sns.barplot(data=prolongued_decelerations_df, x='prolongued_decelerations', y='fetal_health');
plt.xlabel('prolongued_decelerations', fontsize=20, labelpad=20)
plt.ylabel('fetal_health', fontsize=20, labelpad=20);
```

```python
accelerations = df.groupby("accelerations")["fetal_health"].sum().sort_values(ascending = False)
accelerations_df = pd.DataFrame(list(accelerations.items()), columns=['accelerations', 'fetal_health'])
plt.figure(figsize=(15,15))
plt.title('ivmeler ile fetal saÄŸlÄ±ÄŸÄ± oranÄ±')
sns.barplot(data=accelerations_df, x='accelerations', y='fetal_health');
plt.xlabel('accelerations', fontsize=20, labelpad=20)
plt.ylabel('fetal_health', fontsize=20, labelpad=20);
```

```python
histogram_tendency = df.groupby("histogram_tendency")["fetal_health"].sum().sort_values(ascending = False)
histogram_tendency_df = pd.DataFrame(list(histogram_tendency.items()), columns=['histogram_tendency', 'fetal_health'])
plt.figure(figsize=(5,5))
plt.title('histogram eÄŸilimi ile fetal saÄŸlÄ±ÄŸÄ± oranÄ±')
sns.barplot(data=histogram_tendency_df, x='histogram_tendency', y='fetal_health');
plt.xlabel('histogram_tendency', fontsize=20, labelpad=20)
plt.ylabel('fetal_health', fontsize=20, labelpad=20);
```

```python
from warnings import filterwarnings
filterwarnings('ignore')

from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale 
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
y = df["fetal_health"]
X = df.drop(['fetal_health'], axis=1)
X = pd.DataFrame(X)
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size=0.30, 
                                                    random_state=42)
```

```python
from lightgbm import LGBMClassifier
```

```python
lgbm_model = LGBMClassifier().fit(X_train, y_train)
```

```python
y_pred = lgbm_model.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

```python
## Model Tuning
```

```python
lgbm_model
```

```python
lgbm_params = {
        'n_estimators': [100, 500, 1000, 2000],
        'subsample': [0.6, 0.8, 1.0],
        'max_depth': [3, 4, 5,6],
        'learning_rate': [0.1,0.01,0.02,0.05],
        "min_child_samples": [5,10,20]}
```

```python
lgbm = LGBMClassifier()

lgbm_cv_model = GridSearchCV(lgbm, lgbm_params, 
                             cv = 4, 
                             n_jobs = -1, 
                             verbose = 2)
```

```python
lgbm_cv_model.fit(X_train, y_train)
```

```python
lgbm_cv_model.best_params_
```

```python
lgbm = LGBMClassifier(learning_rate = 0.1, 
                       max_depth = 5,
                       subsample = 0.6,
                       n_estimators = 500,
                       min_child_samples = 5,
                       class_weight = 'balanced'  )
```

```python
lgbm_tuned = lgbm.fit(X_train,y_train)
```

```python
y_pred = lgbm_tuned.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```
